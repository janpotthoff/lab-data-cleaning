#### 14. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
selected_users=users[['userId','Reputation','Views','UpVotes','DownVotes']]
selected_users.head()
selected_posts=posts_id[['postId','Score','userId','ViewCount','CommentCount']]
selected_posts.head()
```

#### 15. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merger=selected_users.merge(selected_posts)
merger.head()
```

#### 16. How many missing values do you have in your merged dataframe? On which columns?

#### 17. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
view_counts_null = merger[(merger['ViewCount'].isnull()==True)]
view_counts_null
```

```python
view_counts_null = merger[(merger['ViewCount'].isnull()==False)]
view_counts_null
```

```python
# The Viewcount is often used. It seems like many videos just didn't get views yet. 
# Therefore, we exchange those by 0
merger["ViewCount"] = merger["ViewCount"].fillna(0)
merger
```
